```python
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("rocksamples.csv")
df.head()
```

```python
sns.distplot(a=df['Weight (g)'],kde=False)
```

```python
sns.kdeplot(df['Weight (g)'],shade=True)
#this the smootheed histograph
```

```python
sns.jointplot(x=df['Weight (g)'],y=df['Pristine (%)'],kind='kde')
```

```python
sns.distplot(df['Weight (g)'],kde=False,label='weight') 
sns.distplot(df['Pristine (%)'],kde=False,label='pristine')
plt.title("diffrence")
plt.legend()
```

```python
sns.kdeplot(df['Weight (g)'],shade=True) 
sns.kdeplot(df['Pristine (%)'],shade=True)
plt.title("diffrence")
```
