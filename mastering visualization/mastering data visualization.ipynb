```python
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#loading the data
df=pd.read_csv("rocksamples.csv")
df.head()
```

```python
#setting the width and the height of the figure
plt.figure(figsize=(16,6))
#drwaing the linegraph
sns.lineplot(data=df['Weight (g)'],label='weight of rocks')


```

```python
sns.lineplot(data=df['Pristine (%)'],label="pristine")
```

```python
#plotting a bar chart
plt.figure(figsize=(10,6))
#adding the title
plt.title("graph of weight of stones from the space")
#drawing the bargraph
sns.barplot(x=df.Mission,y=df['Weight (g)'])
plt.ylabel(" weight of rocks")
```

```python
#creating the heatmap
plt.figure(figsize=(14,4))
plt.title("the heatmap of the weight of rocks")
# sns.heatmap(data=df,annot=True)
sns.heatmap(data=df['Weight (g)'], annot=True)
plt.xlabel("weight of rocks")

```

```python
#drawing the scatter plot
sns.scatterplot(x=df['Weight (g)'],y=df['Pristine (%)'],hue=df['Weight (g)'] )
```

```python
sns.regplot(x=df['Weight (g)'],y=df['Pristine (%)'])
```
